## Data Loading

```python
train_df = pd.read_csv("./data/house_train.csv")
test_df = pd.read_csv("./data/house_test.csv")
```

## Data Preprocessing

- set index 'id'

```python
train_df.set_index("Id", inplace=True)
test_df.set_index("Id", inplace=True)
```

- train / test index 저장

```python
train_index = train_df.index
test_index = test_df.index
train_index, test_index
```

- y_train_df 따로 저장

```python
y_train_df = train_df.pop("SalePrice")
```

```python
y_train_df.head(3)
```

- 결측치 처리

```python
train_missval = train_df.isnull().sum() / len(train_df) * 100
train_missval = train_missval[train_missval.values > 0]
train_missval
```

```python
test_missval = test_df.isnull().sum() / len(test_df) * 100
test_missval = test_missval[test_missval > 0]
test_missval
```

- Decision 1 - drop Alley, PoolQC, Fence, MiscFeature
    - 결측치가 너무 많아서 삭제
    - Alley : 골목 접근 유형 (결측치 93.77%)
    - PoolQC : 수영장 품질 (결측치 99.52%)
    - Fence : 울타리 품질 (결측치 80.75%)
    - MiscFeature : 다른 카테고리에서 다루지 않는 기타 기능 -  (결측치 96.30%)

```python
del train_df["Alley"]
del train_df["PoolQC"]
del train_df["Fence"]
del train_df["MiscFeature"]
del test_df["Alley"]
del test_df["PoolQC"]
del test_df["Fence"]
del test_df["MiscFeature"]
```

- 데이터 모판 생성 (y 제외, train + test)

```python
all_df = pd.concat([train_df, test_df], axis = 0)
all_df.tail()
```

```python
all_missval = all_df.isnull().sum() / len(all_df) * 100
all_missval[all_missval.values > 0]
```
