### 결측치 처리

- 데이터 모판 생성 (y 제외, train + test)

```python
all_df = pd.concat([train_df, test_df], axis = 0)
all_df.tail()
```

```python
sns.set(rc = {'figure.figsize' : (10, 8)})
```

```python
all_missval = all_df.isnull().sum() / len(all_df) * 100
all_missval[all_missval.values > 0].sort_values().plot.bar()
```

- **Decision 1** : drop Alley, PoolQC, Fence, MiscFeature
    - 결측치가 너무 많아서 삭제
    - Alley : 골목 접근 유형 (결측치 93.77%)
    - PoolQC : 수영장 품질 (결측치 99.52%)
    - Fence : 울타리 품질 (결측치 80.75%)
    - MiscFeature : 다른 카테고리에서 다루지 않는 기타 기능 -  (결측치 96.30%)

```python
del all_df["Alley"]
del all_df["PoolQC"]
del all_df["Fence"]
del all_df["MiscFeature"]
```

- **Decision 1-1** : 하나의 value가 전체 데이터의 80% 이상을 차지하는 column은 삭제
- 이 경우 의미가 없고, y와의 상관성 또한 낮아질 것이라고 column이라고 생각해서 삭제함.
    - object type
        - Street : 도로 접근 유형 
        - LandContour : 부동산의 평탄도
        - Utilities : 사용 가능한 유틸리티 유형 
        - LandSlope : 재산의 경사
        - Condition1 : 주요 도로 또는 철도와의 근접성
        - Condition2 : 주요 도로 또는 철도와의 근접성 (초가있는 경우)
        - BldgType : 주거 유형
        - RoofMatl : 지붕 재료 
        - ExterCond : 외장재의 현황
        - BsmtCond : 지하실 일반 상태
        - BsmtFinType2 : 두번째 완성 영역의 품질
        - Heating : 난방 유형 
        - CentralAir : 중앙 에어컨
        - Electrical : 전기 시스템
        - Functional : 홈 기능 등급
        - GarageQual : 차고 품질
        - GarageCond : 차고 상태
        - PavedDrive : 포장된 진입로
        - SaleType : 판매 유형
        - SaleCondition : 판매 조건
    - numeric type
        - BsmtFinSF2 : 유형2 마감 면적
        - LowQualFinSF : 모든 층 저품질 마감 면적 
        - BsmtHalfBath : 지하 반 욕실
        - KitchenAbvGr : 주방 수
        - EnclosedPorch : 닫힌 현관 영역
        - 3SsnPorch : 3계절 현관 면적 
        - PoolArea : 수영장 면적 
        - MiscVal : 기타 기능의 $ Value

```python
for col_name in all_df.columns:
    most_freq_value_ratio = (Counter(all_df[col_name]).most_common(1)[0][1] / len(all_df)) * 100
    most_freq_value_ratio = round(most_freq_value_ratio, 2)
    most_freq_value = Counter(all_df[col_name]).most_common(1)[0][0]
    if most_freq_value_ratio > 80:
        print(f"{col_name} is deleted. ({most_freq_value_ratio}% - {most_freq_value})")
        del all_df[col_name]
```

- **Decision 2** : numeric column을 category type column으로 변경
    - MSSubClass : 건물 등급 (16가지)
    - MoSold : 판매 월
    - YrSold : 판매 연도

```python
all_df["MSSubClass"] = all_df["MSSubClass"].astype(str)
all_df["MoSold"] = all_df["MoSold"].astype(str)
all_df["YrSold"] = all_df["YrSold"].astype(str)
```

- numeric column과 object column  구분 저장

```python
numeric_columns = all_df.select_dtypes(["int", "float"]).columns
object_columns = all_df.select_dtypes("object").columns
numeric_columns, object_columns
```

- **Decision 3** : 결측치 채우기

```python
all_missval = all_df.isnull().sum() / len(all_df) * 100
all_missval[all_missval.values > 0]
```

```python
### 가장 흔한 값으로 채우기
# MSZoning : 일반 구역 분류, 가장 흔한 값은 RL
all_df["MSZoning"] = all_df["MSZoning"].fillna("RL")

# Exterior1st : 집 외부 커버, 가장 흔한 값은 VinylSd
all_df["Exterior1st"] = all_df["Exterior1st"].fillna("VinylSd")

# Exterior2nd : 집의 외피, 가장 흔한 값은 VinylSd
all_df["Exterior2nd"] = all_df["Exterior2nd"].fillna("VinylSd")

# KitchenQual : 주방 품질, 가장 흔한 값은 TA
all_df["KitchenQual"] = all_df["KitchenQual"].fillna("TA")


### None 또는 0으로 채우기
# MasVnrType : 벽돌 베니어 유형, None - 베니어 없음
all_df["MasVnrType"] = all_df["MasVnrType"].fillna("None")

# MasVnrArea : 벽돌 베니어 면적, 0 - 베니어 없음
all_df["MasVnrArea"] = all_df["MasVnrArea"].fillna(0)

# 지하실 관련 category 데이터, None - 지하실 없음
# BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1, BsmtType2
for data in ['BsmtQual', 'BsmtExposure', 'BsmtFinType1']:
    all_df[data] = all_df[data].fillna("None")
    
# 지하실 관련 numeric 데이터, 0 - 지하실 없음
# BsmtFinSF1, BsmtFinSF2, BsmtUnfSF, TotalBsmtSF, BsmtFullBath, BsmtHalfBath
for data in ['BsmtFinSF1', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath']:
    all_df[data] = all_df[data].fillna(0)

# FireplaceQu : 벽난로 품질, None - 벽난로 없음
all_df["FireplaceQu"] = all_df["FireplaceQu"].fillna("None")

# 차고 관련 category 데이터, None - 차고 없음
# GarageType, GarageFinish, GarageQual, GarageCond
for data in ['GarageType', 'GarageFinish']:
    all_df[data] = all_df[data].fillna("None")

# 차고 관련 Numeric 데이터, 0 - 차고 없음
# GarageCars, GarageArea
for data in ['GarageCars', 'GarageArea']:
    all_df[data] = all_df[data].fillna(0)

    
### 다른 값으로 채우기
# LotFrontage : 부동산과의 직선거리, 평균으로 채움
all_df["LotFrontage"] = all_df["LotFrontage"].fillna(all_df["LotFrontage"].mean())

# GarageYrBlt : 차고 건설 연도 , YearBuilt(건설 날짜)로 채움
all_df["GarageYrBlt"] = all_df["GarageYrBlt"].fillna(all_df["YearBuilt"])
```

## Feedback
- one-hot encoding은 feature의 unique value 개수가 많은 경우에는 너무 많은 dummie feature를 만들어서 모델의 성능 저하의 요인이 된다.
- 그런 경우에는 다른 방법을 사용하면 좋을 것 같다.
- 예를 들어 Mean Encoding
    - 근데 이걸 할 때 train data에는 없는 value인데 test data에는 있는 value라면 ..
    - 0으로 채우는 것도 가능함

- **Decision 6** : Log Transformation
    - 모델의 성능 향상을 위해 한 쪽으로 치우쳐져 있는 데이터를 정규분포를 따르도록 해 줌.
    - scipy를 사용하여 왜도의 절댓값이 1이 이상인 column은 Log Transformation

```python
from scipy.stats import skew
skewness = all_df[numeric_columns].apply(lambda x: skew(x.dropna())).sort_values(ascending = False)

high_skewness = skewness[abs(skewness) > 1] # 왜도 값이 1 이상인 column만 채택
skew_columns = high_skewness.index
print(high_skewness)
```

```python
fig = plt.figure()
fig.set_size_inches(20, 10)

ax = []
for i, col_name in enumerate(skew_columns):
    ax.append(fig.add_subplot(2,5, i+1))
    x = onehot_all_df[col_name]
    
    ax[i] = sns.distplot(x, bins = 10)
    ax[i].set_title(col_name)
```

- 위와 같이 왜도 값이 1 이상인 column은 log transformation

```python
onehot_all_df[skew_columns] = np.log10(onehot_all_df[skew_columns]+0.5)
log_onehot_all_df = onehot_all_df.copy()
```

```python
fig = plt.figure()
fig.set_size_inches(20, 10)

ax = []
for i, col_name in enumerate(skew_columns):
    ax.append(fig.add_subplot(2,5, i+1))
    x = log_onehot_all_df[col_name]
    
    ax[i] = sns.distplot(x, bins = 10)
    ax[i].set_title(col_name)
```

## Model

- train data를 Cross Validation (5 times)

```python
# 데이터 모판에서 train data와 test data를 분리
train_df = log_onehot_all_df[all_df.index.isin(train_index)]
test_df = log_onehot_all_df[all_df.index.isin(test_index)]
```

```python
# train data를 Cross Validation 하기 위해 X와 y로 지정
X = np.array(train_df)
y = np.array(log_y_train_df)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

kfold = KFold(n_splits = 5, shuffle = True)

mse = []
reg_list = []
X_y_train_list = []
for train_idx, test_idx in kfold.split(X):
    X_train = X[train_idx]
    y_train = y[train_idx]
    
    X_test = X[test_idx]
    y_test = y[test_idx]
    
    reg = LinearRegression()
    reg.fit(X_train, y_train)
    reg_list.append(reg)
    
    y_pred = reg.predict(X_test)
    
    mse.append(mean_squared_error(y_test, y_pred))
    X_y_train_list.append([X_train, y_train])
    
mse
```

## Feedback
- KFold 이렇게 하는 게 아니라
- CrossValidation을 사용해서 하는게 더 편함

- train data를 5 time Cross Validation한 모델 중 mse가 가장 작은 모델을 택해 test data를 예측

```python
best_reg_idx = np.array(mse).argmin()
best_reg = reg_list[best_reg_idx]
best_reg
```

```python
y_test = 10**best_reg.predict(np.array(test_df))
y_test
```

- 함수의 return 값

```python
X_train, y_train = X_y_train_list[best_reg_idx]
X_test = np.array(test_df)
test_id_idx = np.array(test_index)
```

- kaggle에 제출하기 위해 csv 파일 추출

```python
submission_df = pd.DataFrame([test_id_idx, y_test]).T
submission_df.columns = ["Id", "SalePrice"]
submission_df["Id"] = submission_df["Id"].map(np.int32)

submission_df.to_csv("./submission.csv", index = False)
```
