# Matplotlib Graph

```python
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Scatter

* `scatter()` 함수 사용
*  marker: scatter 모양 지정
  * [matplotlib_marker_reference](https://matplotlib.org/3.5.1/gallery/lines_bars_and_markers/marker_reference.html)

```python
data_1 = np.random.rand(512, 2) # rand() : (난수 갯수, 난수쌍)
data_2 = np.random.rand(512, 2)

plt.scatter(data_1[:, 0], data_1[:, 1], c='b', marker='x')
plt.scatter(data_2[:, 0], data_2[:, 1], c='r', marker='^')

plt.show()
```

* `s` : 데이터의 크기를 지정, 데이터의 크기 비교 가능

> scatter를 그릴 땐 x, y, s의 크기가 같아야 함

```python
# bubble chart
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2
plt.scatter(x, y, s = area, c = colors, alpha = 0.5) # alpha : 투명도
plt.show()
```

## Bar chart

* 실제로 쓸 일은 잘 없음

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]

X = np.arange(0,8,2)
```

```python
plt.bar(X + 0.00, data[0], color = 'b', width = 0.50)
plt.bar(X + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50)
plt.xticks(X+0.50, ("A","B","C", "D")) # 가로축 지점마다 이름 지정
plt.show()
```

```python
data = np.array([[5., 25., 50., 20.],
                 [4., 23., 51., 17],
                 [6., 22., 52., 19]])

color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])
X
```

```python
data = np.array([[5., 5., 5., 5.],
                 [4., 23., 51., 17],
                 [6., 22., 52., 19]])

for i in range(3):
    plt.bar(
        X, 
        data[i], 
        bottom = np.sum(data[:i], axis=0), 
        color = color_list[i], 
        label=data_label[i])
# bottom이 아래 그래프의 끝값이자 위 그래프의 시작값
plt.legend()
plt.show()
```

* 가로로 그래프 그리기 가능 : `plt.barh()`

```python
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)

plt.barh(X, women_pop, color = 'r')
plt.barh(X, -men_pop, color = 'b') # 왼쪽 그래프는 음수 값
plt.show()
```

## Histogram

```python
X = np.random.normal(0, 100, 1000) # X ~ N(0, 100)인 X 1000개
plt.hist(X, bins = 100)
plt.show()
```
