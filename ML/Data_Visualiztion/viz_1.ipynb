# Data visualiztion - matplotlib

> `import matplotlib.pyplot as plt`

* 다양한 그래프를 지원하고 pandas와 연동됨.
* pyplot 객체를 사용하여 데이터를 표시.
* pyplot 객체에 여러 그래프들을 쌓은 다음 flush (plt.show())

```python
import matplotlib.pyplot as plt

X = range(100)
Y = range(100)
plt.plot(X, Y) # plot()의 parameter을 알아봐야 함
plt.show()
```

* 최대 단점 : argument를 kwargs로 받음
* 고정된 argument가 없음

* graph는 원래 figure 객체에 생성되기 때문에, pyplot 객체 사용시 기본 figure에 그래프가 그려짐.
* line graph의 경우, (X, Y) 쌍을 sequence 자료형으로 받아서 순서대로 화면에 출력함.

```python
import numpy as np
X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]
```

```python
# 3개의 그래프가 한 figure에 쌓여서 겹쳐서 보여지게 됨.
plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.plot(range(100), range(100))

plt.show()
```

```python
# 값을 입력할 때는 항상 순서대로 정렬해서 입력해야 함.
import random
X_3 = [random.randint(1,100) for _ in range(100)]
Y_3 = [np.cos(value) for value in X_3]

plt.plot(X_3, Y_3)
plt.plot(X_2, Y_2)
plt.show()
```

## Figure & Axes

* matplotlib는 figure 안에 axes로 구성됨
* figure위에 여러 개의 axes를 생성

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(6,3) # 크기 지정 (가로, 세로)
ax_1 = fig.add_subplot(1,2,1) # 1개의 row, 2개의 column 중 1번째 subplot
ax_2 = fig.add_subplot(1,2,2) # 1개의 row, 2개의 column 중 2번째 subplot

ax_1.plot(X_1, Y_1, c = 'b') # color = 'blue' 라는 뜻
ax_2.plot(X_2, Y_2, c = 'g') 
plt.show()
```

## subplots

* subplot의 순서를 grid로 작성

```python
fig = plt.figure()
fig.set_size_inches(8,4) # 사이즈 설정
# plt.style.use('ggplot') # 스타일 적용

ax = []
colors = ['b','g','r','c','m','y','k']
for i in range(1,7):
    ax.append(fig.add_subplot(2,3,i))
    X_1 = np.arange(50)
    Y_1 = np.random.rand(50)
    c = colors[np.random.randint(1, len(colors))]

    ax[i - 1].plot(X_1, Y_1, c=c)
```

## set color

* [matplotlib_api](https://matplotlib.org/2.0.2/api/colors_api.html)
  * 다 외울 수가 없으니 보통은 api를 보고 그림.

* color 속성을 사용
* float : 흑백, rgb color, predefined color 사용

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]

plt.plot(X_1, Y_1, color = "#000000")
plt.plot(X_2, Y_2, color = 'r')

plt.show()
```

## Set linestyle

* [matplotlib_linestyle](https://matplotlib.org/2.0.2/examples/lines_bars_and_markers/linestyles.html)

```python
plt.plot(X_1, Y_1, c = 'b', linestyle = 'dashed')
plt.plot(X_2, Y_2, c = 'r', ls = 'dotted')

plt.show()
```

## Set title

* pyplot에 title함수 사용, figure의 subplot별 입력 가능

```python
plt.plot(X_1, Y_1, c = 'b', linestyle = 'dashed')
plt.plot(X_2, Y_2, c = 'r', ls = 'dotted')

plt.title('Two lines')
plt.show()
```

* latex 타입의 표현도 가능 (수식 표현 가능)

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")

plt.title('$y = ax+b$')
plt.xlabel('x_line') # x축 지정 (xlabel)
plt.ylabel('y_line') # y축 지정 (ylabel)

plt.show()
```

## Set legend

* `legend()` 함수로 범례를 표시함, loc 위치 등 속성 지정
  * 반드시 `plot()`에서 `label='labelname'`을 지정해주어야 함.

```python
plt.plot(X_1, Y_1, c = 'b', linestyle = 'dashed', label='line_1') # label을 지정
plt.plot(X_2, Y_2, c = 'r', ls = 'dotted', label='line_2') # label을 지정
plt.legend(shadow = True, fancybox = True, loc = 'upper right')

plt.title('$y = ax+b$')
plt.xlabel('x_line') # x축 지정 (xlabel)
plt.ylabel('y_line') # y축 지정 (ylabel)

plt.show()
```

## Set grid & xylim

* graph 보조선(격자)을 긋는 grid와 xy축 범위 한계를 지정

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=True, loc="lower right")


plt.grid(True, lw=0.9, ls="--", c="0.9")
# lw : 선의 굵기, ls : 선의 스타일, c : 선의 투명도
plt.xlim(-1000, 2000) # x값의 최소, 최대값
plt.ylim(-1000, 2000) # y값의 최소, 최대값

plt.show()
```

## savefig()

* fig 저장

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=True, loc="lower right")


plt.grid(True, lw=0.9, ls="--", c="0.9")
plt.xlim(-100, 200) 
plt.ylim(-100, 200)

plt.savefig('test.png', c='a')

plt.show()
```

## Set style

* ggplot style을 이용해서 grid를 표현하기도 가능함.
  * [matplotlib_style_sheets_reference](https://matplotlib.org/3.5.1/gallery/style_sheets/style_sheets_reference.html)

```python
plt.style.use('ggplot') # 스타일 적용

plt.plot(X_1, Y_1, c = 'b', linestyle = 'dashed') 
plt.plot(X_2, Y_2, c = 'r', ls = 'dotted') 

plt.show()
```

## Case : Line Graph

* **월별 통화량 데이터 총합 구해보기**

```python
import pandas as pd

df = pd.read_csv("../pandas/data/phone_data.csv")
df.head()
```

```python
result = df.groupby(['month'])['duration'].sum()
result.plot()
```

* 일별 네트워크별 데이터의 총량을 다중 plot으로 작성하라

```python
# date column에서 년월일만 뽑음
def parsedate(date_info):
    import datetime
    
    return date_info[:9]
```

```python
df['day'] = df['date'].map(parsedate)
df.head()
```

```python
# network와 day를 기준으로 duration에 대한 그룹 데이터를 뽑음
result = df.groupby(['network','day'])['duration'].sum().reset_index()
result
```

```python
result['network'].unique().tolist()
```

```python
fig = plt.figure()
fig.set_size_inches(10,10)

network_types = result['network'].unique().tolist()
ax = []
for i in range(1,7):
    ax.append(fig.add_subplot(2,3,i))
    network_name = network_types[i-1]

    plt.title(network_name)

    X_1 = result[result['network'] == network_name]['day']
    Y_1 = result[result['network'] == network_name]['duration']
    ax[i - 1].get_xaxis().set_visible(False) # x축 값이 안보이게
    ax[i - 1].plot(X_1, Y_1)
```
